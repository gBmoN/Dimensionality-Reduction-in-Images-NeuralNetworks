## Python implementation to transform a color image to gray levels (0 to 255) and to binarized (0 and 255), black and white

```python
!git clone 'https://github.com/gBmoN/Dimensionality-Reduction-in-Images-NeuralNetworks.git'
```

```python
import numpy as np
import matplotlib.pyplot as plt 
import math
import skimage
from skimage import filters
```

```python
image_data = '/content/Dimensionality-Reduction-in-Images-NeuralNetworks/img (1).jpg'
image = skimage.io.imread(image_data)
plt.figure(figsize=(10,10))
plt.imshow(image)
```

## Conversion to grayscale

```python
image_r = image[:,:,0]
image_g = image[:,:,1]
image_b = image[:,:,2]
gray_image = np.ceil((image_r + image_g + image_b) / 3)
plt.imshow(gray_image, cmap='gray')

def convert_to_gray(image, luma=False):
    if luma:
        params = [0.299, 0.589, 0.114]
    else:
        params = [0.2125, 0.7154, 0.0721]
    gray_image = np.ceil(np.dot(image[...,:3], params))
    
    gray_image[gray_image > 255] = 255
    
    return gray_image

new_image = convert_to_gray(image, luma=False)
plt.figure(figsize=(10,10))
plt.imshow(new_image, cmap='gray')
```

## Conversion to Binary

```python
threshold1 = filters.threshold_otsu(image)
from IPython.core.pylabtools import figsize
binarized= (new_image > threshold1)*1
plt.figure(figsize=(10,10))
plt.imshow(binarized, cmap = "gray")
```
